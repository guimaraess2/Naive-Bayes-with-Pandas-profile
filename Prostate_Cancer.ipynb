
**Importando as bibliotecas / Importing libraries**

```
from pandas_profiling import ProfileReport
from sklearn.naive_bayes import GaussianNB
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

**Carregando o dataset / Loading the dataset**

```
df= pd.read_csv('Prostate_Cancer.csv')
```

**Armazenando o Profile Report na varável profile1 / Storing the Profile Report in the profile1 variable**

```
profile1 = ProfileReport(df, title='Prostate_Cancer.csv', html={'style' :{'full_width' :True}})
```

**Executação do relatório / Report execution**

```
profile1.to_notebook_iframe()
```

**Contagem dos valores de diagnóstico benigno e maligno / Counting of benign and malign diagnostic values**

```
print(df.groupby('diagnosis_result').size())
```

```
df.head()
```

**Verificando valores nulos /Checking for null values**

```
df1 = df.isnull().sum()
df1
```

**Exclusão de valores nulos / Deleting null values**

```
df1=df.dropna()
df1.isnull().sum()
```

**Separando variáveis de teste e treino / separating test and training variables**

```
x = df1.iloc[:,2:10]
y = df1.iloc[:,1]

```

**Treinando o algoritmo / Training the algorithm** 

```
classificador = GaussianNB()
classificador.fit(x,y)
```

**Inserindo novos valores para: / inserting news values for:**

**radius,	texture,	perimeter,	area,	smoothness,	compactness,	symmetry,** **ractal_dimension**

```
resultado = classificador.predict([[19,	27,	62,	295,	0.102,	0.053,	0.135,	0.069]])
if resultado=='M':
    print('Resultado=Maligno')
else:
    print('Resultado=Benigno')
```
